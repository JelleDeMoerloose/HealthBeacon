```python
import matplotlib.pyplot as plt

# Data
hours = ['1pm', '2pm', '3pm', '4pm', '5pm', '6pm', '7pm', '8pm', '9pm', '10pm', '11pm', '12am', 
         '1am', '2am', '3am', '4am', '5am', '6am', '7am', '8am', '9am', '10am', '11am', '12pm']
calls = [19, 12, 5, 8, 12, 17, 16, 18, 15, 12, 5, 2, 1, 0, 1, 0, 3, 7, 12, 18, 15, 11, 9, 14]

# Plot Styling
plt.figure(figsize=(10, 5))
plt.bar(hours, calls, color='skyblue', edgecolor='blue')  # Customizing bar color and edge color
plt.xlabel('Time', fontsize=14)  # Adding label fontsize
plt.ylabel('Number of Calls', fontsize=14)  # Adding label fontsize
plt.title('Number of Calls in the Past 24 Hours', fontsize=16)  # Adding title fontsize
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotating x-axis labels and setting fontsize
plt.yticks(range(0, 21, 5),fontsize=10)
plt.gca().spines['top'].set_visible(False)  # Removing top spine
plt.gca().spines['right'].set_visible(False)  # Removing right spine
plt.tight_layout()  # Adjusting layout
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adding horizontal grid lines
plt.savefig('images/graph.png')  # Saving the plot with higher dpi and tighter bounding box
plt.show()

```

```python
import matplotlib.pyplot as plt

# Define the waiting times data
waiting_times = [0, 1, 5, 12, 25]

# Create a boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(waiting_times, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue', color='blue'), medianprops=dict(color='green'))

# Add labels and title
plt.xlabel('Waiting Time (minutes)', fontsize=12)
plt.ylabel('')
plt.title('Boxplot of Waiting Times', fontsize=14)

# Customize ticks
plt.xticks(fontsize=10)
plt.yticks([])  # Remove y-axis ticks

# Add gridlines
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the plot
plt.savefig('images/graph2.png')
plt.show()

```
